```python
def isValidAnagram(s,t):
    #Return true,if t is valid anagram of s,false otherwise
    #s="anagram",t="nagaram",True
    count = 0;
    n = len(s)
    for i in s:
        if i in t:
            count +=1
    if count == n:
        return True
            
    return False
print("Valid Anagram %s" % isValidAnagram("anagram", "nagaram") )    
```

```python
def matrixMultiplicationWithoutNumpy(A,B):
    # A is a list of m lists where each list has n elements(m*n)
    # B is a list of n lists where each list has k elements(n*k)
    # return C where C is the matrix multiplication result of A and B(dimension=m*k)
    c = [[0 for j in range(len(B[0]))] for i in range(len(A))]
   
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
               
                c[i][j] += A[i][k] * B[k][j]
            
    
    return c
A = [[12,7,3],
    [4 ,5,6],
    [7 ,8,9]]
# 3x4 matrix
B = [[5,8,1,2],
    [6,7,3,0],
    [4,5,9,1]]
print("Matrix Multiplication without Numpy", matrixMultiplicationWithoutNumpy(A, B))
```

```python
import numpy as np
def matrixMultiplicationWithNumpy(A,B):
    # A is a double-dimensional numpy array(m*n)
    # B is a double-dimensional numpy array(n*k)
    # return C where C is the matrix multiplication result of A and B(dimension=m*k)
    result = np.dot(A, B)
    return result
A = [[12,7,3],
    [4 ,5,6],
    [7 ,8,9]]
# 3x4 matrix
B = [[5,8,1,2],
    [6,7,3,0],
    [4,5,9,1]]
print("Matrix Multiplication with Numpy", matrixMultiplicationWithNumpy(A, B))
```
